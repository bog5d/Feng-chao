### 使用出货明细 并测算经销商排名


```python
import pandas as pd 
import numpy as np

prof = pd.read_excel('/home/bobo/下载/1.1-11.1渠道出货明细.xls',header=2).fillna('/')


#尝试一次性把多个字符替换为空值 
prof['单据类型'] = prof['单据类型'].replace(['往来销售','销售退货','代销结算'],np.nan,regex=True)


prof = prof.fillna(method='pad')

prof = prof.loc[~(prof['开单日期']=='/')]


prof = prof[['单据类型', '开单日期', '货品名称', '数量', '单价', '销售货款']]
prof['小计'] = prof['数量']*prof['单价']
#.groupby('单据类型').sum()[['数量','销售货款']]



#pd.set_option('display.max_columns', 10000, 'display.max_rows', 10000)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
data = prof.groupby('单据类型').sum()[['数量','小计']].sort_values('小计',ascending=False)

data = data.reset_index()

data = data.loc[data['单据类型'].str.contains('-')]


data.reset_index(inplace=True,drop=True)
data.index = data.index+1

#data.head(50)
```

### 使用毛利明细表测算 

```python
#
import pandas as pd 
import numpy as np

prof = pd.read_excel('/home/bobo/下载/毛利明细表 (2).xlsx',header=2).fillna('/')


#尝试一次性把多个字符替换为空值 
prof['单据类型'] = prof['单据类型'].replace(['往来销售','销售退货','代销结算'],np.nan,regex=True)

# 向前填充
prof = prof.fillna(method='pad')
# 去掉小计那些列
prof = prof.loc[~(prof['单据日期']=='/')]

# 只筛选产品，赠品那些不计入进来。
#prof =prof.loc[prof['单据类型'].str.contains('万家乐')]

prof = prof[['单据类型', '单据日期','客户名称', '数量', '销售价', '成本价',
       '销售收入',  '销售毛利', '销售毛利率']]
#prof.head(3)
#基本实现数据清洗。
```

```python
prof.to_csv('销售明细')
```
