```python
import pandas as pd

import numpy as np


df = pd.read_excel('C:/Users/Administrator.USER-20170623BT/Downloads/自1月毛利明细表.xls',header=2)
df.head()

df.columns

# 去掉不必要的列名
df = df[[ '单据编号', '单据日期', '客户名称', '销售价(本)', '成本价(本)',
       '基本数量', '销售收入(本)', '销售成本(本)', '销售毛利(本)', '销售毛利率']]
# 修改列名
df.columns= [[ '型号', '日期','客户名称', '单价', '成本',
       '数量', '销售收入', '销售成本', '销售毛利', '销售毛利率']]


# 把nan值替换，以方便筛选，有nan值不能操作 
df = df.fillna('x')
# 把单据号替换为nan值方便启用fillna向前填充功能
df['型号'].loc[(df.型号.str.contains("SS"))]=np.nan

# 然后在使用向前填充
df = df.fillna(method='ffill')



# 把之前的型号行和小计行去掉，做成原始数据表。 只要删除日期列中带+的即可
df = df.loc[~df.客户名称.str.contains('x')] 
#df.日期.str.contains('+')


#df['date'] = pd.to_datetime(df['date']) #将数据类型转换为日期类型
#df = df.set_index('date') # 将date设置为index

df['日期'] = pd.to_datetime(df['日期']) #将数据类型转换为日期类型
df = df.set_index('日期')
df.head()
```
