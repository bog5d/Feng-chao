```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/home/bobo/下载/收款明细表 (3).xlsx',header=2).fillna('/')
df.head()
```

```python

# df['收款类型'].str.replace('预收款',np.nan)
# series=data['列名'].str.replace(',','-')

# 将预收款字样替换为nan格式，方便向前填充
df['收款类型'] = df['收款类型'].replace('预收款',np.nan,regex=True)
```

```python
df['收款类型'] = df['收款类型'].fillna(method='pad')
```

```python
df = df[['收款类型', '开单日期','预收金额','收款帐户']]
```

```python

df = df.loc[~(df['开单日期']=='/')]
```

```python
 df['业务'],df['客户'] = df['收款类型'].str.split('-',1).str
```

```python
df['开单日期'] = pd.to_datetime(df['开单日期'])
df = df.set_index('开单日期')
```
